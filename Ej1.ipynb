```python
import random
from matplotlib import pyplot as plt
from math import sqrt,cos,sin,pi
import numpy as np
```

```python
#points = np.genfromtxt("datos.txt",delimiter=",")
```

```python
#Creo el Dataset
angulo = random.random()*2*pi
points = []
for j in range(100):
    aux = [random.random()*2-1,random.random()*2-1]
    if (aux[0]*cos(angulo) + aux[1]*sin(angulo) > 0 ):
      aux.append(1)
    else:
      aux.append(-1)
    points.append(aux)
    
x_s = [item[0] for item in points]
y_s = [item[1] for item in points]
c_s = ["r" if item[2] > 0 else "b" for item in points]
```

```python
#Busco el optimizador
th = [0,0]
th_s = [0,0]
for epoch in range(1):
    for j in range(len(points)):
        if  ( ( th[0]*points[j][0] + th[1]*points[j][1] ) * points[j][2] <= 0 ):
            #norm = sqrt(points[j][0]**2 + points[j][1]**2)
            th[0] = th[0] + ( points[j][2] * points[j][0] ) #/ norm
            th[1] = th[1] + ( points[j][2] * points[j][1] ) #/norm

norm = sqrt(th[0] **2 + th[1] **2)
th[0] = th[0] /  norm
th[1] = th[1] /  norm
# print(th)
```

```python
plt.scatter(x_s,y_s,c = c_s)
plt.plot([0,th[0]],[0,th[1]],c ="g")
plt.plot([th[1],-th[1]],[-th[0],th[0]],c ="g")
plt.axis("equal")
plt.show()
```
